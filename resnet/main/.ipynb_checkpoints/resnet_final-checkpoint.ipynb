```python
#Blood cell type prediction using RESNET
from __future__ import print_function
from glob import glob
import numpy as np
import matplotlib.pyplot as plt

import keras
from keras.applications.resnet50 import ResNet50,preprocess_input
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Model
from keras.layers import Dense,Flatten,Input

%matplotlib inline
```

```python
#blood cells images path
train_path = '/home/t1/Desktop/projects/t1/data/blood_cells/dataset2-master/images/TRAIN/'
valid_path = '/home/t1/Desktop/projects/t1/data/blood_cells/dataset2-master/images/TEST/'

# #fruits 360 small dataset
# train_path = '/home/t1/udemy/advanced_cnn/data/fruits-360-small/Training/'
# valid_path = '/home/t1/udemy/advanced_cnn/data/fruits-360-small/Validation/'

# #fruits 360 large dataset
# train_path = '/home/t1/udemy/advanced_cnn/data/fruits-360/Training/'
# valid_path = '/home/t1/udemy/advanced_cnn/data/fruits-360/Test/'
```

```python
train_images = glob(train_path+'*/*.jp*g')
valid_images = glob(valid_path+'*/*.jp*g')
folders = glob(train_path + '*')
```

```python
folders
```

```python
#random image
random_image = np.random.choice(train_images)
image_label = random_image.split('/')[-2]
temp_image = plt.imread(random_image)
plt.imshow(temp_image)
print('Image label : ',image_label)
```

```python
Image_Size = [224,224]
n_epochs = 5
batch_size = 32
```

```python
#creating the model
resnet = ResNet50(input_shape=Image_Size + [3],weights='imagenet',include_top=False)
```

```python
#dont train the existing layers
for layer in resnet.layers:
    layer.trainable = False
```

```python
x = Flatten()(resnet.output)
predictions = Dense(len(folders),activation='softmax')(x)
```

```python
resnet_model = Model(inputs=resnet.inputs,outputs=predictions)
resnet_model.summary()
```

```python
resnet_model.compile(loss='categorical_crossentropy',optimizer = 'adam' , metrics = ['accuracy'])
```

```python
#Image data generator
gen = ImageDataGenerator(
    rotation_range = 20,
    width_shift_range = 0.1,
    height_shift_range = 0.1,
    shear_range = 0.1,
    zoom_range = 0.1,
    horizontal_flip = True,
    vertical_flip = True,
    preprocessing_function = preprocess_input
)
```

```python
train_gen = gen.flow_from_directory(train_path,target_size=Image_Size,shuffle= True ,batch_size=batch_size)
```

```python
print(train_gen.class_indices)
```

```python
labels = [None]*len(train_gen.class_indices)
for label,i in train_gen.class_indices.items():
    labels[i] = label
    
labels
```

```python
# for x,y in train_gen:
#     print('Min : ',x[0].min(),'Max : ',x[0].max())
#     plt.imshow(x[0])
#     print('Label : ',labels[np.argmax(y[0])])
#     break
```

```python
valid_gen = gen.flow_from_directory(
    valid_path,
    target_size = Image_Size,
    batch_size = batch_size
)
```

```python
r = resnet_model.fit_generator(train_gen,
                               validation_data=valid_gen,
                               epochs=n_epochs,
                               steps_per_epoch=len(train_images)//batch_size,
                               validation_steps = len(valid_images)//batch_size
                              )
```

```python
from sklearn.metrics import confusion_matrix
def get_confusion_matrix(data_path,N):
    print('Generating confusion matrix',N)
    preds = []
    targets = []
    i = 0
    data_gen = gen.flow_from_directory(data_path,
                                    target_size=Image_Size,
                                    shuffle = False,
                                    batch_size = batch_size*2)
  
    for x,y in data_gen:
        i += 1
        if i % 50 == 0:
            print(i)
        p = resnet_model.predict(x)
        p = np.argmax(p,axis=1)
        y = np.argmax(y,axis=1)
        preds = np.concatenate((preds,p))
        targets = np.concatenate((targets,y))
        if len(targets) >= N:
            break
   
    cm = confusion_matrix(targets,preds)
    return cm
```

```python
train_cm = get_confusion_matrix(train_path,len(train_images))
print(train_cm)
valid_cm = get_confusion_matrix(valid_path,len(valid_images))
print(valid_cm)
```

```python
#plotting losses
plt.plot(r.history['loss'],label='train loss')
plt.plot(r.history['val_loss'],label='valid loss')
plt.legend()
```

```python
#plotting accuracies
plt.plot(r.history['acc'],label = 'train acc')
plt.plot(r.history['val_acc'],label = 'valid acc')
plt.legend()
```
